```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)+1
test_size = int(len(data)*.1)+1
```

```python
print(train_size)
print(val_size)
print(test_size)
```

```python
train_size + val_size + test_size
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
#Conv2D - Convolutional Neural Network
#2D convolutional layer (i.e., spatial convolution over images)
#Dropout is used for regularization
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16,(3,3),1,activation='relu',input_shape=(256,256,3)))
model.add(MaxPooling2D())
#here, number of filters = 16

model.add(Conv2D(32,(3,3),1,activation='relu'))
model.add(MaxPooling2D())
#here, number of filters = 32


model.add(Conv2D(16,(3,3),1,activation='relu'))
model.add(MaxPooling2D())
#here, number of filters = 16

model.add(Flatten())

model.add(Dense(256,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
#tf.optimizers.
```

```python
model.compile('adam',loss=tf.losses.BinaryCrossentropy(),metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

```python
# Evaluate the model on the test set
test_predictions = model.predict(test)
test_predictions_binary = (test_predictions > 0.5).astype(int)

# Extract labels from the test set
test_labels = []
for batch in test.as_numpy_iterator():
    _, labels = batch
    test_labels.extend(labels)

# Create a confusion matrix for the test set
test_conf_matrix = confusion_matrix(test_labels, test_predictions_binary)

# Display the confusion matrix using a heatmap
class_labels = ['Happy', 'Sad']
df_cm = pd.DataFrame(test_conf_matrix, index=class_labels, columns=class_labels)

plt.figure(figsize=(8, 6))
sns.heatmap(df_cm, annot=True, cmap='Blues', fmt='g')
plt.title('Test Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
